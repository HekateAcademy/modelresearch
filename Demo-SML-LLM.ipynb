#Import libraries

```python
#import libraries
import torch
from transformers import pipeline
import time
from groq import Groq
```

#Small Language Model

##TinyLlama-1.1B

```python

# Ghi lại thời gian bắt đầu
start_time = time.time()

# Khởi tạo pipeline cho mô hình "TinyLlama-1.1B-Chat-v1.0" với kiểu dữ liệu torch.bfloat16 và tự động phân bổ thiết bị (device_map="auto")
pipe = pipeline("text-generation", model="TinyLlama/TinyLlama-1.1B-Chat-v1.0", torch_dtype=torch.bfloat16, device_map="auto")

# Định nghĩa các tin nhắn đầu vào cho chatbot
messages = [
    {
        "role": "system",  # Vai trò của người tạo ra hệ thống chatbot
        "content": "You are a friendly chatbot who always responds in the style of a pirate",  # Nội dung mô tả phong cách trả lời của chatbot
    },
    {"role": "user", "content": "The key to life is"},  # Nội dung người dùng yêu cầu
]

# Áp dụng template chat và chuyển đổi tin nhắn thành prompt cho mô hình
prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

# Sinh văn bản từ mô hình với các tham số: max_new_tokens=256, do_sample=True, temperature=0.7, top_k=50, top_p=0.95
outputs = pipe(prompt, max_new_tokens=256, do_sample=True, temperature=0.7, top_k=50, top_p=0.95)

# Ghi lại thời gian kết thúc
end_time = time.time()

# Lấy văn bản đã sinh từ đầu ra, cắt bỏ phần trước dấu "<|assistant|>"
tinyllama11b_output = outputs[0]["generated_text"].split("<|assistant|>")[1]

# Tính toán thời gian đã trôi qua
time_taken_tiny11b = end_time - start_time

# In ra văn bản sinh được
print(tinyllama11b_output)

# In ra thời gian thực hiện
print(f"Time taken:{time_taken_tiny11b} seconds")
```

##Llama-3.2-3B

```python
# Ghi lại thời gian bắt đầu
start_time = time.time()

# Định nghĩa model ID cho mô hình Llama-3.2-3B
model_id = "meta-llama/Llama-3.2-3B"

# Khởi tạo pipeline cho mô hình "Llama-3.2-3B" với kiểu dữ liệu torch.bfloat16 và tự động phân bổ thiết bị (device_map="auto")
pipe = pipeline(
    "text-generation",  # Chỉ định tác vụ sinh văn bản
    model=model_id,  # Đưa vào model ID đã xác định
    torch_dtype=torch.bfloat16,  # Sử dụng kiểu dữ liệu torch.bfloat16 cho phép tính toán hiệu quả hơn
    device_map="auto"  # Tự động phân bổ mô hình sang thiết bị phù hợp (GPU/CPU)
)

# Sinh văn bản từ mô hình với đầu vào "The key to life is"
llama32_3b = pipe("The key to life is")

# Ghi lại thời gian kết thúc
end_time = time.time()

# Tính toán thời gian đã trôi qua
time_taken_3b = end_time - start_time

# In ra kết quả sinh văn bản từ mô hình
print(llama32_3b)

# In ra thời gian thực hiện quá trình sinh văn bản
print(f"Time taken:{time_taken_3b} seconds")
```

#Large Language Model

##Llama-3-70B : Hugging Face

Kết quả báo lỗi vì không đủ bộ nhớ

```python
# Xác định model ID cho mô hình "Meta-Llama-3-70B-Instruct"
model_id = "meta-llama/Meta-Llama-3-70B-Instruct"

# Khởi tạo pipeline cho mô hình "Meta-Llama-3-70B-Instruct" với các tham số: kiểu dữ liệu torch.bfloat16 và tự động phân bổ thiết bị
pipeline = transformers.pipeline(
    "text-generation",  # Chỉ định tác vụ sinh văn bản
    model=model_id,  # Đưa vào model ID đã xác định
    model_kwargs={"torch_dtype": torch.bfloat16},  # Cung cấp kiểu dữ liệu cho mô hình (torch.bfloat16)
    device_map="auto",  # Tự động phân bổ mô hình sang thiết bị (GPU/CPU)
)

# Định nghĩa các tin nhắn đầu vào cho chatbot, bao gồm vai trò của hệ thống và người dùng
messages = [
    {"role": "system", "content": "You are a pirate chatbot who always responds in pirate speak!"},  # Mô tả chatbot với phong cách trả lời của cướp biển
    {"role": "user", "content": "The key to life is"},  # Nội dung người dùng yêu cầu
]

# Định nghĩa các terminators (dấu kết thúc) để dừng quá trình sinh văn bản
terminators = [
    pipeline.tokenizer.eos_token_id,  # Sử dụng mã token EOS (End Of Sentence) của tokenizer
    pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")  # Mã token cho một terminator khác
]

# Sinh văn bản từ mô hình với các tham số: tối đa 256 token mới, các terminators đã xác định, sampling và các tham số sampling khác
outputs = pipeline(
    messages,
    max_new_tokens=256,  # Số lượng token tối đa có thể sinh ra
    eos_token_id=terminators,  # Các token kết thúc để dừng quá trình sinh
    do_sample=True,  # Cho phép sampling trong quá trình sinh văn bản
    temperature=0.6,  # Điều chỉnh mức độ ngẫu nhiên (temperature) trong quá trình sinh
    top_p=0.9,  # Chỉ giữ lại top-p phần trăm các token có xác suất cao nhất
)

# In ra ký tự cuối cùng trong văn bản sinh được
print(outputs[0]["generated_text"][-1])  # In ký tự cuối cùng của văn bản được sinh từ mô hình
```

##Llama-3-70B: Groq

Để lấy API Key từ Groq, thực hiện theo các bước sau:

1. Truy cập trang [https://console.groq.com/keys](https://console.groq.com/keys).
2. Đăng nhập vào tài khoản Groq.
3. Tạo và sao chép API key từ trang quản lý API.


```python
import os

os.environ["GROQ_API_KEY"] = "YOUR API KEY"  # Replace with your actual API key
```

```python
# Ghi lại thời gian bắt đầu
start_time = time.time()

# Khởi tạo client Groq với khóa API từ biến môi trường
client = Groq(
    api_key=os.environ.get("GROQ_API_KEY"),  # Lấy khóa API từ biến môi trường (cần thiết để truy cập API của Groq)
)

# Gửi yêu cầu tạo một câu trả lời từ mô hình chat của Groq
chat_completion = client.chat.completions.create(
    messages=[  # Danh sách tin nhắn người dùng gửi
        {
            "role": "user",  # Vai trò của người gửi tin nhắn (người dùng)
            "content": "The key to life is",  # Nội dung câu hỏi của người dùng
        }
    ],
    model="llama-3.3-70b-versatile",  # Chỉ định mô hình được sử dụng
)

# Ghi lại thời gian kết thúc
end_time = time.time()

# Lấy nội dung câu trả lời của mô hình từ lựa chọn đầu tiên
llama70b = chat_completion.choices[0].message.content

# Tính toán thời gian đã trôi qua
time_taken_70b = end_time - start_time

# In ra câu trả lời từ mô hình
print(llama70b)

# In ra thời gian thực hiện quá trình trả lời
print(f"Time taken:{time_taken_70b} seconds")
```
